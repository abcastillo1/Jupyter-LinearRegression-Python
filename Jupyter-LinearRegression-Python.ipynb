```python
from scipy import stats
```

```python
# Impotar biblioteca numpy para poder crear vectores y matrices
```

```python
import numpy as np
```

## Generar los datos

```python
# Se crea un array que contiene 30 números aleatorios
```

```python
rng = np.random.default_rng ()
x = rng.random(30)
y = 2.3 * x + rng.random(30)
x
y
```

## Cálculo de la regresión lineal

```python
slope,intercept,r,p,std_err=stats.linregress(x,y)
```

```python
# La función retorna los valores de: 
# pendiente, intersección, coeficiente de correlación,
# prueba de hipótesis, error estándar.
```

```python
slope,intercept,r,p,std_err
```

## Función Auxiliar

```python
# La pendiente e intersección calculadas para devolver
# un nuevo valor.
```

```python
def fitted_value(x):
    return slope * x + intercept
```

## Valores ajustados para el eje "Y"

```python
new_values=list(map(fitted_value,x))
new_values
```

## Graficar el modelo 

```python
# Los datos se encuentran representados en la gráfica 
```

```python
plt.plot(x,y,'o',label='Dato Original')
plt.plot(x,new_values,'r',label='Línea Ajustada')
plt.legend()
```
